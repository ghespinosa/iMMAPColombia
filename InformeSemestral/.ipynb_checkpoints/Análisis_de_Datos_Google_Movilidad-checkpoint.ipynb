## PlotMaker Function

```python
def PlotMaker(Department,PlaceCategory):

  if PlaceCategory in cat_.keys():
    place_cat = cat_.get(PlaceCategory)
  else:
    print('Error: ' + PlaceCategory + ' is not an accepted category.')

  # Subset PlaceCategory data
  col_place_cat = col_mobi[['sub_region_1','date',place_cat]]

  # Subset Department data
  col_place_cat_dpt = col_place_cat[col_place_cat['sub_region_1']==Department].reset_index(drop=True)

  # Create Google Mobility Report

  x_values = [] #generate x and y values for the plot
  for date in col_place_cat_dpt.date:
    x_values.append(dt.datetime.strptime(date,"%Y-%m-%d").date()) #convert to datetime
                                                        
  y_values = col_place_cat_dpt[place_cat].tolist()

  ax = plt.gca() #get axes
  ax.set_title(Department) #+ " " + PlaceCategory)

  #ax.legend(loc='center right',bbox_to_anchor=(0.15, 0.6))

  #formatter = mdates.DateFormatter("%Y-%m-%d") #format as dates
  #ax.xaxis.set_major_formatter(formatter) 

  #locator = mdates.DayLocator() #set locator
  #ax.xaxis.set_major_locator(locator)

  plt.rcParams["figure.figsize"] = [30, 5]

  plt.plot(x_values, y_values) #plot values
```

```python
# Plot one department's plots all on the same line

dpt = 'North Santander'

for cat in cat_.keys():
  PlotMaker(dpt,cat)
```

```python
# Plot one department's plots all on the same line

dpt = 'Bogota'

for cat in cat_.keys():
  PlotMaker(dpt,cat)
```

```python
# Plot ALL plots

for dpt in dpt_:
  for cat in cat_.keys():
    if (type(dpt) == str):
      PlotMaker(dpt,cat)
```
